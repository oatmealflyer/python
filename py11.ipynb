## 데이터 프레임 정렬

```python
import numpy as np
import pandas as pd
print("DataFrame: ")
data = {
    'col1':[2,1,9,8,7,4],
    'col2':['A','A','B',np.nan,'D','C'],
    'col3':[0,1,9,4,2,3]
}
index = [1,3,2,5,0,4]
df = pd.DataFrame(data,index=index)
df
```

```python
# sort_index axis=0(행) 을 축으로 정렬
df = df.sort_index(axis=0)
df
```

```python
# sort_index axis=1(열) 을 축으로 정렬, 내림차순
df = df.sort_index(axis=1, ascending=False)
df
```

```python
# sort_value 1
sorted_df1 = df.sort_values('col1', ascending=True)
sorted_df1
```

```python
# sort_value 2
sorted_df2 = df.sort_values(['col2','col1'],ascending=[True,False])
sorted_df2
```

## 데이터 분석용 함수

```python
# 데이터 프레임의 index와 columns
import numpy as np
import pandas as pd
data = {
    'korean': [50,60,70,30],
    'math':[30,np.nan,40,20]
}
index = ['a','b','c','d']
df = pd.DataFrame(data,index=index)
print(df.index)
print(df.columns)
df
```

```python
# count() 함수
print(df.count(axis=0)) # axis = 0 기본값(생략가능)
print(df.count(axis=1))
```

```python
# max(), min(), sum(), mean()
print(df.max())
print(df.max(axis=1))
print(df.min())
print(df.min(axis=1))
print(df.sum())
print(df.sum(axis=1))
print(df.mean())
print(df.mean(axis=1))
print(df.mean(axis=1,skipna=False)) # NaN값 보이게 하기
```

```python
# 결측값 대체 fillna
avg = df['math'].mean()
df['math']=df['math'].fillna(avg) 
df
```

## 그룹으로 묶기 (groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key'   : ['A','A','A','A','B','C'],
    'data1':[1,2,3,1,2,3,],
    'data2':[4,4,6,0,6,1],
    # 'data3':['안녕','a','b','하세요','pple','3']
    })
print("DataFrame:")
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum()#.reset_index()
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'max','data2':'sum'})
```

```python
# filter 메서드
t=list(range(1,11))
result=[]
def is_even(x):
    return True if x%2 ==0 else False
for i in t:
    if is_even(i):
        result.append(i)
print(result)
```

```python
df.groupby('key').mean()
```

```python
def filter_mean(x):
    return x['data2'].mean() > 3
df.groupby('key').filter(filter_mean)
# filter_mean(df.groupby('key'))
```

```python
file_path="./ref/행정구역시도_성연령별취업자_20250304.csv"
ori=pd.read_csv(file_path,encoding="EUC-KR")
ori.head()
```

```python
ex_data =ori.copy() #복사 
ex_data.columns = ['region','year','age','gender','value'] #칼럼이름 일괄교체 
# ex_data.groupby()
ex_data.loc[ex_data['region']=='세종특별자치시','value'] =ex_data.loc[ex_data['region']=='세종특별자치시','value'].replace('-',0) 
ex_data
```

```python
ex_data['value'] = ex_data['value'].astype(int)
ex1_data = ex_data.groupby('year')['value'].sum().reset_index(name="total")
```
