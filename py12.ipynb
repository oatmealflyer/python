```python
file_path = "./ref/애완용품_판매매체별_상품군별거래액.csv"
ori = pd.read_csv(file_path,encoding="EUC-KR") #utf-8
ori.columns = ['category','date','total','internet','mobile']
ori.tail()
```

```python
print("수정전")
print(ori['date'].tail())
print("-"*15)
print("수정후")
ori['date'] = ori['date'].map(lambda x: x.rstrip(' p)'))
print(ori['date'].tail())
```
